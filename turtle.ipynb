```python
import turtle
turtle.clear()
turtle.pencolor('red')
turtle.pensize(2)
turtle.circle(100)
turtle.clear()
turtle.setposition(0,0)
turtle.forward(100)
turtle.left(90)
turtle.forward(100)
turtle.left(90)
turtle.forward(100)
turtle.left(90)
turtle.forward(100)
turtle.left(90)
import turtle
turtle.clear()
turtle.colormode(255)
turtle.color(150,100,100)
turtle.pendown()
turtle.fd(200)
turtle.left(170)
turtle.fd(200)
turtle.left(170)
turtle.fd(200)
turtle.left(170)
turtle.fd(200)
turtle.left(170)

# turtle.exitonclick()
# turtle.done()
from turtle import *

turtle.clear()
turtle.setposition(0,0)
color('red', 'yellow')
begin_fill()
while (1):
	forward(200)
	left(170)
	if abs(pos()) < 1:
		break
end_fill()
done()
# mainloop()
import turtle
turtle.clear()
turtle.penup()
turtle.goto(0,0)
turtle.begin_fill()
turtle.color('blue')
# print (turtle.pos())
turtle.circle(100)
turtle.end_fill()

turtle.colormode(255)
turtle.color(190, 200, 150)
turtle.pencolor(240, 160, 80)

turtle.penup()
turtle.goto(-40,100)
turtle.begin_fill()
turtle.circle(50)
turtle.end_fill()

# turtle.setposition(-130,100)
# turtle.right(45)
# turtle.fd(200)
# turtle.setposition(150,100)
turtle.penup()
turtle.goto(140,100)
turtle.begin_fill()
turtle.circle(50)
turtle.end_fill()
#!/usr/bin/env python3
"""       turtle-example-suite:

            tdemo_paint.py

A simple  event-driven paint program

- left mouse button moves turtle
- middle mouse button changes color
- right mouse button toggles between pen up
(no line drawn when the turtle moves) and
pen down (line is drawn). If pen up follows
at least two pen-down moves, the polygon that
includes the starting point is filled.
 -------------------------------------------
 Play around by clicking into the canvas
 using all three mouse buttons.
 -------------------------------------------
          To exit press STOP button
 -------------------------------------------
"""
from turtle import *

def switchupdown(x=0, y=0):
    if pen()["pendown"]:
        end_fill()
        up()
    else:
        down()
        begin_fill()

def changecolor(x=0, y=0):
    global colors
    colors = colors[1:]+colors[:1]
    color(colors[0])

def main():
    global colors
    shape("circle")
    resizemode("user")
    shapesize(.5)
    width(3)
    colors=["red", "green", "blue", "yellow"]
    color(colors[0])
    switchupdown()
    onscreenclick(goto,1)
    onscreenclick(changecolor,2)
    onscreenclick(switchupdown,3)
    return "EVENTLOOP"

if __name__ == "__main__":
    msg = main()
    print(msg)
    mainloop()
```
